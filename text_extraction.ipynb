```python
import PyPDF2
import os

pdf_folder = '/content/gdrive/MyDrive/Maestria/10k_predictor/data/raw/unstructured'

pdf_files = os.listdir(pdf_folder)
pdf_files.sort()

pdf_texts = []

for file in pdf_files:
    if file.endswith('.pdf'):
        file_path = os.path.join(pdf_folder, file)
        with open(file_path, 'rb') as pdf:
            pdf_reader = PyPDF2.PdfReader(pdf)
            text = ''
            for page in pdf_reader.pages:
                text += page.extract_text()
            pdf_texts.append(text)

#print(pdf_texts)
```

```python
pdf2 = pdf_texts
```

```python
for i in range(len(pdf2)):
    pdf2[i] = pdf2[i].replace('\n', ' ')
    pdf2[i] = pdf2[i].replace('\xa0', '')
    pdf2[i] = pdf2[i].replace(',', '')
```

```python
csvpath = '/content/gdrive/MyDrive/Maestria/10k_predictor/data/raw/structured/RawDataset.csv'
df = pd.read_csv(csvpath)
df = df.rename(columns={'Ticker': 'ticker','Name': 'name','Co Des': 'co_des','WACC Economic Value Added': 'eva','WACC EVA Spd':'eva_spd','Mkt Cap': 'mkt_cap','Num': 'id', 'Desempeño': 'performance'})
df = df[['id','ticker','name','eva','eva_spd','mkt_cap','performance']]
n = len(pdf2)
df = df.head(n)
df['texto'] = pdf2
```
